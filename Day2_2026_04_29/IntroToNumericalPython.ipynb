A package is a collection of modules that are not necessarily included in the base Python distribution. The name **numpy** is abbreviation for _numerical Python_. To access **numpy** we must first **import** it

```python
x = [3, 4, 5]
y = [1, 2, 3]
x+y
```

```python
import numpy as np
```

In the previous line, we named numpy as module np; an abbreviation for easier referencing. We use the _np.array()_ to define x and y, which are one-dimensional arrays, i.e. vectors.

```python
x = np.array([3, 4, 5])
y = np.array([1, 2, 3])
```

Since x and y have been defined using np.array(), we get a sensible result when we add them together. It is completely differ from define x and y without using numpy

In numpy, matrices are typically represented as two-dimensional arrays, and vectors as one-dimensional arrays. We can create two-dimensional arrays as follows:

```python
z = np.array([[1, 2], [3, 4], [34,2]])
z
```

The object z has several _attributes_, or _associated object_. To access an attribute of z, we type _z.attribute_ where we replace _attribute_ with the name of the attribute. For instance, we can access the _ndim_ attribute of z as follows.

```python
z.ndim
```

The output indicates that z is a two-dimensional array. Similarly, _z.dtype_ is the datatype attribute of the object z. This indicates that z comprises 64-bit integers:

```python
z.dtype
```

Why is z comprised in integer? This is because we created z by passing in exclusively integers to the np.array() function. If we had passed in any decimals, then we would have obtained any array of floating point numbers (i.e. real-valued numbers)

```python
np.array([[1, 2],[3.0, 4]]).dtype
```

The array z is two-dimensional. We can find out the numbers of rows and columns by looking at its _shape_ attribute.

```python
z.shape
```

What is a method? A method is a function that associate with an object. For instance, given an array x, the expression of _x.sum()_ sums all of its elements, using the **sum()** method for arrays. The call **x.sum()** automatically provides x as the first argument to its **sum()** method.

```python
x = np.array([1,2.2,3,4])
x.sum()
```

We could also sum the elements of x by passing in x as an argument to the **np.sum()** function.

```python
x = np.array([1,2.2,3,4])
np.sum(x)
```

As another example, the **reshape()** method return a new array with the same elements of x, but a difference shape. We do this by passing in a **tuple** in our call to **reshape()**, in this case **(2, 3)**. This tuple specifies that we would like to create a two-dimensional array with 2 rows and 3 columns.

Like lists, tuples represent a sequence of objects. Why do we need more than one way to create a sequence? There are a few differences between tuples and lists, but perhaps the most important is that elements of a tuple **cannot** be modified, whereas elements of a list can be.

```python
x = np.array([1,2,6,3,4,7])
print("beginning x:\n", x)
x_reshape = x.reshape(2, 3)
print("reshaped x:\n", x_reshape)
```

The previous outputs reveals that numpy arrays are specified as a sequence of rows. This is called row-major ordering, as opposed to column-major ordering. Python (and hence numpy) uses 0-based indexing. This means that to access the top left element of **x_reshape**, we type in **x_reshape[0,0]**

```python
x_reshape[0, 0]
```

Similarly, x_reshape[1,2] yields the element in the second row and the third column of _x_reshape_

```python
x_reshape[1,2]
```

Similarly, x[2] yields the third entry of x. Now, let's modify the top left element of x_reshape. To our surprise, we discover that the first element of x has been modified as well!

```python
print('x before we modify x_reshape:\n',x)
print('x_reshape before we modify x_reshape:\n',x_reshape)
x_reshape[0,0] = 5
print('x_reshape after we modify its top left element:\n',x_reshape)
print('x after we modify top left element of x_reshape:\n',x)
```

Modifying x_reshape also modified x because the two objects occupy the same space in memory. We just saw that we can modify an element of an array. Can we also modify a tuple? It turns out that we **cannot** - and trying to do so introduces an _exception_, or error.

```python
my_tuple = (1, 2, 3)
my_tuple[0] = 2
```

```python
my_list = [1, 2, 3]
my_list[0] = -1
my_list
```

We now briefly mention some attributes of arrays that will come in handy. An array's _shape_ attribute contains its dimension; this is always a tuple. The _ndim_ attribute yields the number of dimensions, and _T_ provides its transpose.

```python
print('x_reshape:\n',x_reshape)
print('x_reshape shape:\n',x_reshape.shape)
print('x_reshape dimensions:\n',x_reshape.ndim)
print('x_reshape transpose:\n',x_reshape.T)
```

Notice that the three individual outputs **(2,3); 2; and array [[5 3] [2 4] [6 7]]** are themselves output as a tuple. We will often want to apply functions to arrays. For instance, we can compute the square root of the entries using the **np.sqrt()** function:

```python
print(np.sqrt(x_reshape))
```

We can also square the elements

```python
np.square(x_reshape)
```

We will often want to generate random data. The **np.random.normal()** function generates a vector of random normal variables. We can learn more about this function by looking at the help page, via a call to **np.random.normal?**. The first line of the help page reads **normal(loc=0.0, scale=1.0, size=None)**. This _signature_ line tells us that the function's arguments are **loc, scale, and size**. These are keyword arguments, which mean that when they are passed into the function, they can be referred to by name (in any order). By default, this function will generate random normal variable(s), with mean (**loc**) 0 and standard deviation (**scale**) 1, furthermore, a single random variables will be generated unless the argument to **size** is changed. We now generate 50 independent random variables from _N(0,1)_ distribution.

```python
x = np.random.normal(size=50)
x
```

We create an array _y_ by adding an independent _N(50,1)_ random variables to each element of _x_

```python
y = x + np.random.normal(loc=50, scale=1, size=50)
```

The **np.corrcoef()** function computes the _correlation matrix_ between x and y. The off-diagonal elements give the correlation between x and y

```python
np.corrcoef(x,y)
```

We can use np.mean(), np.var(), and np.std() function to compute mean, variance, and standard deviation of arrays. These functions are also available as methods on the arrays.

```python
x = np.array([[2,3],[4,5],[6,7],[8,9]])
x
```

```python
np.mean(x), x.mean()
```

```python
np.var(x), x.var(), np.mean((x-x.mean())**2)
```

```python
np.std(x), x.std(), np.sqrt(np.var(x))
```

The np.mean(), np.var(), and np.std() can also be applied to the rows and columns of a matrix. To see this, we construct a 10x3 matrix of N(0,1) random variables, and consider computing its row sums.

```python
rng = np.random.default_rng(3)
```

```python
X = rng.standard_normal((10,3))
X
```

```python
X.mean(), X.std()
```

Since arrays are row-major ordered, the first axis aka first col, i.e. axis=0, refers to its rows. We pass this argument into the mean() method for object x.

```python
X.mean(axis=0)
```

The following yields the same result.

```python
X.mean(0)
```
